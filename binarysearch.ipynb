My first solution :

```python
nums = list(map(int, input().split()))
target = int(input())

def search(nums, target):
    left, right = 0, len(nums) - 1

    while left <= right:
        mid = (left + right) // 2

        if nums[mid] == target:
            return mid
        elif nums[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return -1

print(search(nums, target))
```

my second solution

```python
nums = list(map(int, input().split()))
target = int(input())

def binary(nums, target, left, right):
    if left > right:
        return -1

    mid = (left + right) // 2

    if nums[mid] == target:
        return mid
    elif nums[mid] < target:
        return binary(nums, target, mid + 1, right)
    else:
        return binary(nums, target, left, mid - 1)

print(binary(nums, target, 0, len(nums) - 1))
```

third solution

```python
import bisect

nums = list(map(int, input().split()))
target = int(input())

def search(nums, target):
    idx = bisect.bisect_left(nums, target)

    if idx < len(nums) and nums[idx] == target:
        return idx
    return -1

print(search(nums, target))
```
